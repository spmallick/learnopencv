```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from mtcnn import MTCNN

%matplotlib inline
```

```python
# 1. reading the image and then converting the image from BGR to RGB 
# 2. then displaying the image
```

```python
#1
image = cv2.imread("Images1/ppl.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
#2
plt.imshow(image)
```

```python
# MTCNN library is used for detecting faces
```

### MTCNN
MTCNN is an abbrevation for Multi-Task Cascaded Convolutional Neural Network. The MTCNN is popular because it achieved then state-of-the-art results on a range of benchmark datasets, and because it is capable of also recognizing other facial features such as eyes and mouth, called landmark detection.

```python
mtcnn = MTCNN()
```

```python
# 1.detecting the faces using the image
# 2.and then taking a single face and drawing the bounding box
# 3.after that displaying the faces with the obtained bounding boxes
```

```python
#1
faces = mtcnn.detect_faces(image)
draw_rect = np.copy(image)
#2
for face in faces:
    x, y, w, h = face['box']
    draw_rect = cv2.rectangle(draw_rect, (x, y), (x+w, y+h), (0, 255, 64), 2)
#3
plt.imshow(draw_rect)
```

```python
# 1.now using numpy library taking the box and applying the blurr effect on that part only
# 2.then saving the blurred image
# 3.after that displaying it
```

```python
#1
faces = mtcnn.detect_faces(image)
blurred = np.copy(image)

for face in faces:
    x, y, w, h = face['box']
    blurred[y:y+h, x:x+h] = cv2.blur(blurred[y:y+h, x:x+h], (20, 20))
#2
plt.imsave("Images1/Blurred_Image.png", blurred)    
#3
plt.imshow(blurred)
```

Practical applications of face blurring and anonymization include:

- Privacy and identity protection in public/private areas
- Protecting children online (i.e., blur faces of minors in uploaded photos)
- Photo journalism and news reporting (e.g., blur faces of people who did not sign a waiver form)
- Dataset curation and distribution (e.g., anonymize individuals in dataset)
- Blurring the faces of those who don't sign a release then attempts to avoid the issue by making sure people aren't identifiable.

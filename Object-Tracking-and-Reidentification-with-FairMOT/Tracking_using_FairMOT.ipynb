```python
# Import dependencies.
from __future__ import absolute_import
from __future__ import division
from __future__ import print_function
import _init_paths
import logging
import os
import os.path as osp
from opts import opts
from tracking_utils.utils import mkdir_if_missing
from tracking_utils.log import logger
import datasets.dataset.jde as datasets
from track import eval_seq
import shutil
logger.setLevel(logging.INFO)
```

```python
def demo(opt):
    result_root = opt.output_root if opt.output_root != '' else '.'
    mkdir_if_missing(result_root)

    # Setting file names.
    slash = opt.input_video.rindex('/')+1
    ext = opt.input_video.rindex('.')
    res_vid = opt.input_video[slash:ext] + '.mp4'
    logger.info('Starting tracking...')
    dataloader = datasets.LoadVideo(opt.input_video, opt.img_size)
    res_txt = opt.input_video[slash:ext] + '.txt'
    result_filename = os.path.join(result_root, res_txt)
    frame_rate = dataloader.frame_rate

    # Purging and recreating the frame directory.
    op_dir = osp.join(result_root, 'frame')
    if osp.exists(op_dir) and osp.isdir(op_dir):
        shutil.rmtree(op_dir)
    frame_dir = None if opt.output_format == 'text' else osp.join(result_root, 'frame')

    # Calling the function to start tracking.
    eval_seq(opt, dataloader, 'mot', result_filename,
             save_dir=frame_dir, show_image=False, frame_rate=frame_rate,
             use_cuda=opt.gpus!=[-1])

    # Combine the frames into the video.
    if opt.output_format == 'video':
        output_video_path = osp.join(result_root, res_vid)
        cmd_str = 'ffmpeg -f image2 -i {}/%05d.jpg -b 5000k -c:v mpeg4 {}'.format(osp.join(result_root, 'frame'), output_video_path)
        os.system(cmd_str)
```

```python
# Pass arguments
arg = ['mot',
       '--load_model=/content/FairMOT/models/fairmot_dla34.pth',
       '--arch=dla_34',
       '--input-video=/content/FairMOT/videos/MOT16-03.mp4',
       '--track_buffer=30',
       '--output-root=/content/FairMOT/tests',
       '--conf_thres=0.4']

# Call the function.
os.environ['CUDA_VISIBLE_DEVICES'] = '0'
opt = opts().init(args = arg)
demo(opt)
```

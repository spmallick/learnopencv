```python
import os
import random
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras.datasets import boston_housing
```

The first step is to provide determinism support. Thus, in order to get reproducible model results we need to configure all random numbers generators.

```python
# settings for reproducibility
seed = 42
random.seed(seed)
tf.random.set_seed(seed)
np.random.seed(seed)

os.environ['TF_DETERMINISTIC_OPS'] = '1'
```

```python
(X_train, Y_train), (X_test, Y_test) = boston_housing.load_data()
```

```python
nFeatures = X_train.shape[1]

model = Sequential()
model.add(Dense(1, input_shape=(nFeatures,), kernel_initializer='uniform'))
model.add(Activation('linear'))

model.compile(optimizer='rmsprop',  loss='mse', metrics=['mse', 'mae'])

model.fit(X_train, Y_train, batch_size=4, epochs=1000)
```

```python
model.summary()

model.evaluate(X_test, Y_test, verbose=True)
```

```python
Y_pred = model.predict(X_test)
```
